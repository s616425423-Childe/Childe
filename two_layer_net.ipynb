In particular, an issue commonly seen with activation functions is getting zero (or close to zero) gradient flow during backpropagation. Which of the following activation functions have this problem? If you consider these functions in the one dimensional case, what types of input would lead to this behaviour?
1. Sigmoid
2. ReLU
3. Leaky ReLU

$\color{blue}{\textit Your Answer:}$ *Sigmoid and ReLU.

当输入为较大的正值或负值时，Sigmoid 激活函数产生的梯度会非常接近于零。在这些区域，Sigmoid 函数处于饱和状态，其输出值接近 1 或 0；因此，其导数变得极小，导致几乎没有梯度能反向流经该单元。

当输入为负值时，ReLU 产生的梯度为零。在此区域，ReLU 的输出和导数均为零。如果 ReLU 单元持续接收负值输入，它可能会停止学习；这种情况有时被称为“死亡 ReLU”（dead ReLU）。

Leaky ReLU 通常不存在这个问题，因为它在输入为负值时具有一个较小的非零斜率。因此，即使输入为负值，梯度仍能继续流动。*
