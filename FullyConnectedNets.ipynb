Now, try to use a five-layer network with 100 units on each layer to overfit on 50 training examples. Again, you will have to adjust the learning rate and weight initialization scale, but you should be able to achieve 100% training accuracy within 20 epochs.

````python
# TODO: Use a five-layer Net to overfit 50 training examples by
# tweaking just the learning rate and initialization scale.
candidate_configs = [
    # weight_scale, learning_rate
    (0.070, 0.008),
    (0.070, 0.010),
    (0.070, 0.005),
    (0.065, 0.010),
    (0.065, 0.015),
    (0.075, 0.005),
    (0.075, 0.008),
    (0.080, 0.003),
]

search_results = []
best_model = None
best_solver = None
best_train_acc = -1.0

for weight_scale, learning_rate in candidate_configs:
    np.random.seed(231)

    candidate_model = FullyConnectedNet(
        [100, 100, 100, 100],
        weight_scale=weight_scale,
        reg=0.0,
        dtype=np.float64
    )

    initial_loss, _ = candidate_model.loss(
        small_data["X_train"],
        small_data["y_train"]
    )

    # 初始损失异常时跳过
    if not np.isfinite(initial_loss):
        print(
            f"scale={weight_scale:.3f}, "
            f"lr={learning_rate:.3f}: "
            "initial loss 非有限，跳过"
        )
        continue

    candidate_solver = Solver(
        candidate_model,
        small_data,
        print_every=1000,
        num_epochs=20,
        batch_size=25,
        update_rule="sgd",
        optim_config={
            "learning_rate": learning_rate
        },
        lr_decay=1.0,
        verbose=False
    )

    candidate_solver.train()

    loss_history = np.asarray(
        candidate_solver.loss_history
    )

    is_finite = np.isfinite(loss_history).all()

    if is_finite:
        max_train_acc = max(
            candidate_solver.train_acc_history
        )
        final_loss = loss_history[-1]
    else:
        max_train_acc = 0.0
        final_loss = np.inf

    result = {
        "weight_scale": weight_scale,
        "learning_rate": learning_rate,
        "initial_loss": initial_loss,
        "final_loss": final_loss,
        "max_train_acc": max_train_acc,
        "finite": is_finite,
    }

    search_results.append(result)

    print(
        "scale=%.3f lr=%.3f "
        "initial_loss=%.4f final_loss=%.4f "
        "max_train_acc=%.3f finite=%s"
        % (
            weight_scale,
            learning_rate,
            initial_loss,
            final_loss,
            max_train_acc,
            is_finite,
        )
    )

    if (
        is_finite
        and max_train_acc > best_train_acc
    ):
        best_train_acc = max_train_acc
        best_model = candidate_model
        best_solver = candidate_solver

print("\n最佳训练准确率：", best_train_acc)

plt.plot(solver.loss_history)
plt.title('Training loss history')
plt.xlabel('Iteration')
plt.ylabel('Training loss')
plt.grid(linestyle='--', linewidth=0.5)
plt.show()
````

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
[在 AdaGrad 算法中，“缓存”（cache）存储的是梯度的平方和。由于梯度的平方是非负的，该缓存值会随时间推移不断增大。随着缓存值的增大，参数更新公式中的分母也随之变大，导致有效学习率逐渐降低。最终，参数的更新幅度可能变得极小，从而使学习过程变慢甚至实际上停止。

Adam 算法不太容易出现这一特定问题，因为它的二阶矩估计采用的是指数移动平均，而非无界累加和。旧的梯度平方值会逐渐衰减，而不会永久保留在缓存中。此外，Adam 还使用了一阶矩估计和偏差校正机制。尽管在接近收敛或超参数设置不当的情况下，Adam 的更新幅度也可能变小，但它并不存在 AdaGrad 那种单调增长的累加器结构。]
